```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
X_train = pd.read_csv('../data/X_train.csv').values
X_test = pd.read_csv('../data/X_test.csv').values
y_train = pd.read_csv('../data/y_train.csv').values.ravel()
y_test = pd.read_csv('../data/y_test.csv').values.ravel()
```

```python
Logitsic_model = LogisticRegression(max_iter=1000)
Logitsic_model.fit(X_train,y_train)
Kn_model = KNeighborsClassifier(n_neighbors=5)
Kn_model.fit(X_train,y_train)

Logitsic_model_pred = Logitsic_model.predict(X_test)
Kn_model_pred = Kn_model.predict(X_test)
```

```python
LR_accuracy = accuracy_score(y_test, Logitsic_model_pred)
Kn_accuracy = accuracy_score(y_test, Kn_model_pred)

LR_c_matrix = confusion_matrix(y_test, Logitsic_model_pred)
Kn_c_matrix = confusion_matrix(y_test, Kn_model_pred)

LR_c_report = classification_report(y_test,Logitsic_model_pred)
Kn_c_report = classification_report(y_test,Kn_model_pred)

print(f'accuracy for LR : {LR_accuracy}')
print(f'accuracy for KN : {Kn_accuracy}')
print(f'Confusion Matrix for LR : {LR_c_matrix}')
print(f'Confusion Matrix for KN : {Kn_c_matrix}')
print(f'Classification report for LR : {LR_c_report}')
print(f'Classification report for KN: {Kn_c_report}')
```
