```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

```python
train_df = pd.read_csv('../data/mitbih_train.csv' , header=None)
test_df = pd.read_csv('../data/mitbih_test.csv', header=None)

print("train shape:", train_df.shape)
print("test shape:", test_df.shape)
```

```python
# Adding Column names

train_df.columns = [f'feature_{i}' for i in range(187)] + ['label']
test_df.columns = [f'feature_{i}' for i in range(187)] + ['label']

train_df.head()
```

```python
print(train_df['label'].value_counts())
sbn.countplot(x='label', data=train_df)
plt.title('Class Distribution in Training Data')

plt.show()
```

```python
sample = train_df[train_df['label'] == 0.0].iloc[0, :187]
plt.figure(figsize=(9, 4))
plt.plot(sample.values)
plt.title('Sample Normal ECG Beat')
plt.xlabel('Time Steps')
plt.ylabel('Amplitude')
plt.show()
```
